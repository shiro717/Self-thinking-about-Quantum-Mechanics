```python
import math
import matplotlib.pyplot as plt

n = int(input("Please enter an integer: "))

energy_level = []
degeneracy = []

for a in range(1, n + 1):
    b = 0
    for i in range(0, int(math.sqrt(a)) + 1):
        for j in range(i, int(math.sqrt(a - i ** 2)) + 1):
            for k in range(j, int(math.sqrt(a - i ** 2 - j ** 2)) + 1):
                if a == i ** 2 + j ** 2 + k ** 2:
                    if i == j == k:
                        b += 1
                    elif i == j or j == k or i == k:
                        b += 3
                    else:
                        b += 6
    energy_level.append(a)
    degeneracy.append(b)
    print(f"Degeneracy of energy level {a} is {b}")
```

```python
plt.scatter(energy_level, degeneracy, color='blue', s=10, alpha=0.8)

plt.title("Relation Between Energy Level and Degeneracy")
plt.xlabel("Energy Level")
plt.ylabel("Degeneracy")
plt.show
```
